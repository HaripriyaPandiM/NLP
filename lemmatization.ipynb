```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
text = "The children are playing and eating in the garden"
words = word_tokenize(text)
stop_words = set(stopwords.words("english"))
filtered_words = [w for w in words if w.lower() not in stop_words]
lemmatizer = WordNetLemmatizer()
lemmatized_words = [
    lemmatizer.lemmatize(w, 'v') if w.endswith('ing')
    else lemmatizer.lemmatize(w)
    for w in filtered_words
]
print("Original Text:", text)
print("After Stopword Removal:", filtered_words)
print("After Lemmatization:", lemmatized_words)
```
